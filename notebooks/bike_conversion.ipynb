```python
# this notebook is a proof of concept for converting truck points in bike clusters into bike points
```

```python
import pandas as pd
import folium
```

```python
bikepoints = pd.read_csv("../data/bike_service_pts_galv.csv")
bikepoints.head()
```

```python
truckpoints = pd.read_csv("../data/truck_service_pts_galv.csv")
truckpoints.head()
```

```python
def add_markers(f_map, points, color='blue'):
    """
    given a folium map, route data (includes location names), and a color (str),
    draw markers on the given map

    Parameters:
        f_map : folium map
        route_data : pd.DataFrame
        color : str (e.g., "red", "blue")
    """
    
    icon_size= 100
    for i in range(len(points)):
        y, x = points[["Latitude", "Longitude"]].iloc[i]
        folium.Marker((y,x), popup=points.iloc[i]['Name'], icon=folium.Icon(color=color)).add_to(f_map)
        icon_size = 40
    return None
```

```python
map = folium.Map(location = [29.30135, -94.7977], tiles ='OpenStreetMap', zoom_start=11)
add_markers(map, truckpoints, 'blue')
add_markers(map, bikepoints, 'red')
```

```python
truth_df = pd.read_csv("../data/indoor_outdoor_pts_galv.csv")
truth_df.head()
```

```python
truth_dist = pd.read_csv("../data/indoor_outdoor_distances_galv.csv")
truth_dist
```

```python
truth_df["pickup_type"] == 'Bike'
```

```python
bike_rows = truth_dist.iloc[truth_df[truth_df["pickup_type"] == 'Bike'].index.to_list()]
drop_cols = [str(i) for i in truth_df[truth_df["pickup_type"] == 'Bike'].index]
bike_to_truck_dists = bike_rows.drop(columns = drop_cols)
bike_to_truck_dists
```

```python
converts = []
for col in bike_to_truck_dists.columns:
    if min(bike_to_truck_dists[col]) <= 500:
        converts.append(int(col))
```

```python
converted_df = truth_df.copy()
for i in converts:
    converted_df.at[i, "pickup_type"] = 'Bike'
```

```python
converted_df.iloc[converts]["pickup_type"]
```

```python
truckpoints2 = converted_df[converted_df["pickup_type"] == 'Truck']
bikepoints2 = converted_df[converted_df["pickup_type"] == 'Bike']
```

```python
map = folium.Map(location = [29.30135, -94.7977], tiles ='OpenStreetMap', zoom_start=11)
add_markers(map, truckpoints2, 'blue')
add_markers(map, bikepoints2, 'red')
map
```

```python
truckpoints2.to_csv(path_or_buf="../data/converted_truck_service_pts_galv.csv")
bikepoints2.to_csv(path_or_buf="../data/converted_bike_service_pts_galv.csv")
```
