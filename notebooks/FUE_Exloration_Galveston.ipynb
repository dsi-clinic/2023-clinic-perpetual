# FUE's of Galveston Visualization

### Importing Libraries

```python
import pandas as pd
import folium
```

### Visualization

```python
# Every entry is a different restaurant
FEU_df = pd.read_csv("../data/FUE_Galveston.csv")
FEU_df.head()
```

```python
# We want to plot the visualization so all we need are locations of each point
FEU_coordinates = FEU_df[['Latitude', 'Longitude']]
FEU_coordinates

# turning values into list pairs
CoordinateList = FEU_coordinates.values.tolist()
# googled galveston city center
GalvestonCenter= [29.3013, -94.7977]

# built map with galveston at the centerlink used: 
# https://towardsdatascience.com/creating-a-simple-map-with-folium-and-python-4c083abfff94
map = folium.Map(location=GalvestonCenter, zoom_start=12)
for point in range(0, len(CoordinateList)):
    folium.Marker(CoordinateList[point], 
                popup=FEU_df['Name'][point]).add_to(map)
display(map)
```
