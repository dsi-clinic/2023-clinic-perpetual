# create_trucks_only_dfs
This notebook will subset the original galveston indoor+outdoor location and distance matrix dataframes to truck service only. 

```python
import pandas as pd
```

### Subset Galveston indoor + outdoor locations df to trucks only

```python
#read in the galveston locations df:
#       this is the original full_df we made by concatonating 
#       moody gardens + indoor df + outdoor df together
location_df = pd.read_csv("../data/indoor_outdoor_pts_galv.csv")
```

```python
#subset to truck only rows
truck_only_df = location_df.loc[(location_df.loc[:, "pickup_type"]=="Truck"), :]
```

```python
#reset index
truck_only_df.reset_index(inplace=True)
#set the original indeces for future reference
truck_only_df = truck_only_df.rename(columns={"index": "Original_Index"})
```

```python
truck_only_df.head()
```

```python
truck_only_df.to_csv("../data/truck_service_pts_galv.csv", index=False)
```

### Subset Galveston indoor + outdoor distance matrix to trucks only

```python
#import distance matrix
distance_matrix = pd.read_csv("../data/indoor_outdoor_distances_galv.csv")
```

```python
#get list of index locations for all bike service locations
bike_indices = location_df.loc[(location_df.loc[:, "pickup_type"]==
                              "Bike"), :].index.values.tolist()
```

```python
#get the column names for all bike service locations
columns_for_bikes = [str(i) for i in bike_indices]
```

```python
#drop all rows of index with bike service
distance_matrix = distance_matrix.drop(bike_indices, axis=0)
#drop all columns with bike service
distance_matrix = distance_matrix.drop(columns_for_bikes, axis=1)
```

```python
#reset index and columns
distance_matrix.reset_index(inplace=True, drop=True)

num_truck = len(location_df) - len(bike_indices)
distance_matrix.columns = [str(i) for i in range(num_truck)]

distance_matrix
```

```python
distance_matrix.to_csv("../data/truck_distances_galv.csv", index=False)
```
